# Top Countries by GDP in 2023

```python
latest_year = df.loc[2023]
top_countries = latest_year.sort_values(ascending=True).head()
top_countries
```

```python
top_countries.mean()
```

```python
top_countries.median()
```

```python
top_countries.mode()
```

# Average Annual GDP Growth Rate

```python
growth_rates = df.pct_change().mean() * 100  # Convert to percentage
growth_rates.sort_values(ascending=False)
growth_rates
```

```python
growth_rates.value_counts().plot(kind='pie', figsize=(12, 8))
```

# Total GDP of All Countries Over Time

```python
total_gdp = df.sum(axis=1)
total_gdp.plot(figsize=(12, 6), title='Total GDP of European Countries (1960_2023)')
plt.ylabel('Total GDP (USD)')
plt.show()
```
